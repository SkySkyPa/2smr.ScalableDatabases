```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
x = np.linspace(1, 10, 21)
y = x ** 2
```

```python
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Sample Plot')
plt.xlim()
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x,y, 'purple')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('X to Y')
plt.xlim()

plt.subplot(1, 2, 2)
plt.plot(y,x,'p')
plt.xlabel('Y Value')
plt.ylabel('X Value')
plt.title('Y to X')
plt.xlim()

plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x,y,'--x',color='blue',)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('X to Y')
plt.xlim()

plt.subplot(2, 1, 2)
plt.plot(y,x, 'g', linestyle = 'dotted')
plt.xlabel('Y Value')
plt.ylabel('X Value')
plt.title('Y to X')
plt.xlim()

plt.show()
```

```python
x1 = [2, 1, 3, 6, 8, 9]
y1 = [4, 3, 1, 2, 9, 10]
```

```python
plt.subplot(2, 2, 1)
plt.plot(x,y,'--x',color='red',)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('X to Y')
plt.xlim()

plt.subplot(2, 2, 2)
plt.plot(x,y,'-',color='green',)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('X to Y')
plt.xlim()

plt.subplot(2, 2, 3)
plt.plot(x,y,color='purple',)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('X to Y')
plt.xlim()

plt.subplot(2, 2, 4)
plt.plot(x1,y1,linestyle='dotted')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('X to Y')

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8,8))

axes[0,0].plot(x,y)
axes[0,0].set_title('First')

fig.savefig('plots.png', dpi=500)
```

```python
x = np.linspace(1, 15, 3)
y = x**2
z = x**3
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, label= 'x vs y', color = 'purple', alpha = 0.5)
axes.plot(x,z, label='x vs z')
axes.set_xticks([0,2,4,6,8,10])
axes.set_xticklabels(['Zero', 'Two', 'four', 'six', 'eight', 'ten'])
axes.set_xlabel('X lab')
axes.set_ylabel('Y lab')
axes.set_title('Object Oriented Interface')
axes.legend(loc = 0)
plt.show
```
